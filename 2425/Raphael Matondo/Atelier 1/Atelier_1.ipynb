**Atelier avec SQL**

```python
import sqlite3

conn = sqlite3.connect('/content/supermarket_sales.db')

c = conn.cursor()

c.execute('''
CREATE TABLE IF NOT EXISTS supermarket_sales (
    "Invoice ID" INTEGER,
    "Branch" TEXT,
    "City" TEXT,
    "Customer type" TEXT,
    "Gender" TEXT,
    "Product line" TEXT,
    "Unit price" REAL,
    "Quantity" INTEGER,
    "Tax 5%" REAL,
    "Total" REAL,
    "Date" DATE,
    "Time" TIME,
    "Payment" TEXT,
    "cogs" REAL,
    "gross margin percentage" REAL,
    "gross income" REAL,
    "Rating" REAL
)
''')

```

```python
# Insertion des données du DataFrame dans la table
data[["Invoice ID", "Branch", "City", "Customer type", "Gender", "Product line",
      "Unit price", "Quantity", "Tax 5%", "Total", "Date", "Time", "Payment",
      "cogs", "gross margin percentage", "gross income", "Rating"]].to_sql('supermarket_sales', conn, if_exists='replace', index=False)
```

```python
# Requête SQL
query = 'SELECT * FROM supermarket_sales'

# Exécuter la requête et charger les résultats dans un DataFrame
df = pd.read_sql_query(query, conn)

# Affichage des résultats
print(df.head())  # Affiche les premières lignes du DataFrame
```

```python
# Requête SQL pour sélectionner des colonnes spécifiques
query = 'SELECT "Invoice ID", "Branch", "City", "Total", "Date" FROM supermarket_sales'

# Charger les résultats dans un DataFrame
df = pd.read_sql_query(query, conn)

# Affichage du DataFrame
print(df)
```

```python
#Filtrer par ville
c.execute("SELECT * FROM supermarket_sales WHERE City = 'Yangon';")
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Filtrer par plage de dates
c.execute('''SELECT * FROM supermarket_sales
             WHERE Date BETWEEN '01-01-2019' AND '31-12-2019';''')
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Filtrer par type de client et genre
c.execute('''SELECT * FROM supermarket_sales
             WHERE "Customer type" = 'Member' AND Gender = 'Female';''')
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Total des ventes par succursale
c.execute('''SELECT Branch, SUM(Total) AS Total_Ventes
             FROM supermarket_sales
             GROUP BY Branch;''')
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Moyenne par ligne de produits
c.execute('''SELECT "Product line", AVG(Rating) AS Moyenne_Évaluation
             FROM supermarket_sales
             GROUP BY "Product line";''')
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Revenus et taxes totales par ville
c.execute('''SELECT City, SUM("gross income") AS Total_Revenus, SUM("Tax 5%") AS Total_Taxes
             FROM supermarket_sales
             GROUP BY City;''')
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Produits les plus vendus
c.execute('''SELECT "Product line", SUM(Total) AS Total_Ventes
             FROM supermarket_sales
             GROUP BY "Product line"
             ORDER BY Total_Ventes DESC
             LIMIT 5;''')
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Résumé des ventes mensuelles
c.execute('''SELECT strftime('%Y-%m', Date) AS Mois, SUM(Total) AS Ventes_Mensuelles
             FROM supermarket_sales
             GROUP BY Mois
             ORDER BY Mois;''')
rows = c.fetchall()
for row in rows:
    print(row)
```

```python
#Mettre à jour une colonne
c.execute('''UPDATE supermarket_sales
             SET "gross margin percentage" = 15
             WHERE "Product line" = 'Électronique';''')
conn.commit()
print("Mise à jour effectuée avec succès.")
```

```python
#Supprimer les données obsolètes
c.execute('DELETE FROM supermarket_sales WHERE Date < "2023-01-01";')
conn.commit()
print("Données obsolètes supprimées avec succès.")
```
